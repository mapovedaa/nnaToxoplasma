```python
import pandas as pd
import numpy as np 
```

```python
datos = pd.read_csv('data/data.csv')
datos = datos.drop(['Genes'],axis=1)
datos.head()
```

```python
datos.describe(include ='all')
```

```python
datos = datos.T
from   sklearn.feature_selection    import VarianceThreshold
umbral = 0.7
sel    = VarianceThreshold(threshold=(umbral * (1 - umbral)))
sel.fit_transform(datos)
datos  = datos[datos.columns[sel.get_support(indices=True)]]
datos = datos.T
datos.info()
```
